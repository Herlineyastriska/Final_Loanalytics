```python
#Duplicated Data
df.duplicated().sum()
```

```python
#Handle Outliers
from scipy import stats
print(f'Jml baris sebelum outlier di filter: {len(df)}')

filter = np.array([True] * len(df))
for col in ['Age']:
    zscore = abs(stats.zscore(df[col])) #hitung absolut zscore
    filter = (zscore<3) & filter #keep yang kurang dari 3 absolute z-score nya

df = df[filter]

print(f'Jml baris setelah outlier di filter: {len(df)}')
```

```python
print(f'Jml baris sebelum outlier di filter: {len(df)}')

filter = np.array([True] * len(df))
for col in ['Age']:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    low_limit = Q1 - (1.5 * IQR)
    high_limit = Q3 + (1.5 * IQR)
    
    filter = ((df[col] >= low_limit) & (df[col] <= high_limit))

df = df[filter]
print(f'Jml baris setelah outlier di filter: {len(df)}')
```

```python
#Feature Encoding
mapping_status = {'single' : 0, 'married' : 1,}
df['Married/Single'] = df['Married/Single'].map(mapping_status)
df.head()
```

```python
#Feature Encoding
mapping_car = {
    'no' : 0,
    'yes' : 1
}
df['Car_Ownership'] = df['Car_Ownership'].map(mapping_car)
df.head()
```

```python
#Membuat fitur baru dengan age
def segment(x):
    if x['Age'] >=21 and x['Age'] <= 30:
        segment = 'Young Adults'
    elif x['Age'] <= 45:
        segment = 'Middle-aged Adults'
    else :
        segment = 'Old-aged Adults'
    return segment
```

```python
df_new['age_group'] = df_new.apply(lambda x: segment(x), axis = 1)
df_new.head()
```

```python
#Membuat fitur baru dengan income
def segment(x):
    if x['Income'] >=1.031000e+04 and x['Income'] <=2.503015e+06:
        segment = 'Low Income'
    elif x['Income'] <=7.477502e+06	:
        segment = 'Medium Income'
    else :
        segment = 'High Income'
    return segment
```

```python
df_new['income_group'] = df_new.apply(lambda x: segment(x), axis = 1)
df_new.head()
```

```python
#Membuat fitur baru dengan current job years group
def segment(x):
    if x['CURRENT_JOB_YRS'] >=0 and x['CURRENT_JOB_YRS'] <=3:
        segment = 'Entry Level'
    elif x['CURRENT_JOB_YRS'] <= 9	:
        segment = 'Mid-Senior Level'
    else :
        segment = 'Manager Level'
    return segment
```

```python
df_new['job_status_group'] = df_new.apply(lambda x: segment(x), axis = 1)
df_new.head()
```
